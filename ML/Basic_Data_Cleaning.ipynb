#**Basic Data Cleaning**

In Machine learning, data cleaning is basically the “process of fixing or removing data that is inaccurate, duplicated, or outside the scope of your research question” (https://datascience.cancer.gov/training/learn-data-science/clean-data-basics).

We will leverage Pandas and NumPy libraries to perform some data cleaning steps.


In this tutorial, you will learn:

* Identify and remove zero-variance predictors (columns with a single unique value) that provide no discriminatory power for machine learning models.
* Remove redundant features where information is perfectly duplicated or derived from other existing columns.
* How to identify and remove rows that contain duplicate observations.

### Case 1-Columns Containing a Single Value
 In machine learning, columns that contain the same value across all observations are known as zero-variance predictors. While they may seem harmless, they introduce several technical and structural issues into the modeling pipeline.

Thus, the primary problem is the lack of information content:
* Zero Discriminatory Power: A feature that never changes cannot help a model distinguish between different classes or predict a numerical target

* Constant Input: Mathematically, if x is a constant, the derivative of the loss function with respect to that feature's weight will be zero, meaning the model cannot "learn" anything from it.

Another problem is the computational inefficiency as every additional column will increase memory and processing power, as well as  dimensionality of the data.

**Recommendation for Data Preparation**

The standard practice is to programmatically identify these columns using df.nunique() and drop them immediately during the cleaning phase.

Removing single-value columns is a foundational step in Feature Selection that ensures your model focuses only on variables that actually capture biological or clinical variance.


```python
# First, let's identify which columns contain a single value.
#summarize the number of unique values for each column using pandas, using .nunique
# load the dataset
import pandas as pd
df = pd.read_csv('messy2.csv', header=None)
# summarize the number of unique values in each column using nunique()
print("Shape of messy data: ", df.shape)
print("Column\t#Unique values ")
print(df.describe())
#We can print unique values specifically
print(df.nunique())
```

We can see that column 6 with index 5 is the only one with a single value. Checking on the imported file the value is 4. This column should be removed as it won't influence the prediction.

```python
#Alternatively using functions as you learned in function
for i in df.columns:
  unique_val= df[i].nunique()
  print(f'Count of column in {i}:', unique_val)
```

```python
# Alternatively, when you don't know which ones should be dropped or would like to do it systematically
# load the dataset
df = pd.read_csv('messy2.csv', header=None)
print(df.shape)
# get number of unique values for each column using nunique
counts = df.nunique()
# record columns to delete: This is a list comprehension in Python. It creates a new list, to_del, from the indices i of the counts list where the value v is equal to 1. In other words, it's finding the positions of all elements in counts that are equal to 1.
to_del = [i for i,v in enumerate(counts) if v == 1]
print(to_del)
# drop useless columns
df.drop(to_del, axis=1, inplace=True)
print(df.shape)
df.head()

```

### Case 3- Redundant observations (rows)

Redundant observations can lead to data leakage.

**The Problem:** If you have technical replicates (identical or near-identical rows from the same patient) and you perform a random train-test split, one replicate might end up in the Training Set and another in the Testing Set.

**The Consequence:** The model will "cheat." Instead of learning the biological patterns of a disease, it simply "remembers" the specific values for that individual patient. This results in artificially high accuracy during testing that will vanish when the model is applied to truly new patients in the real world

In most standard ML pipelines, you should not keep raw replicates if:
   You are using a simple Random Split: As mentioned

*   You are using a simple Random Split of training and test sets
*   One class is over-represented: If "Healthy" samples have 5 replicates each but "Disease" samples only have 1, the model will be biased toward the "Healthy" features simply because it saw them more often, not because they are biologically more important.
*  You are calculating feature importance: Duplicates can inflate the perceived importance of features that are unique to the duplicated samples.

There are cases where it is OK to keep replicates, but special strategies are needed for training the model. This is out of scope of this course.

```python
# Use the .duplicated() function to isolate these cases.
# Identify all rows where every feature value is identical.
# Inspect the duplicates using keep=False.The keep=False argument marks all duplicates as True. It will allow you to inspect all instances of a duplicate. So, df2[df2.duplicated(keep=False)] returns all the duplicate rows in the DataFrame.
#Determine if these represent computational artifacts (to be deleted) or biological replicates (to be handled via grouped cross-validation).

#To print duplicates related to particular column information, then use df.duplicated(subset['col1', 'col2'], keep=False)
duplicate_rows = df2[df2.duplicated(keep=False)]
print(duplicate_rows)
```
