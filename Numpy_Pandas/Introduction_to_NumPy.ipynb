#The Basics
NumPy’s main object is the homogeneous multidimensional array (ndarray). It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers. In NumPy dimensions are called *axes*.

##An example NumPy ndarray


```python
import numpy as np

# 'arange()' returns evenly spaced values within a given interval.
# Values are generated within the half-open interval [start, stop)
# (in other words, the interval including start but excluding stop).

a = np.arange(15)
a
```

```python
# 'reshape()' gives a new shape to an array without changing its data.
a = a.reshape(3, 5)
a
```

###**ndarray.ndim**
the number of axes (dimensions) of the array.



```python
a.ndim
```

###**ndarray.shape**
the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the number of axes, ndim.

```python
a.shape
```

###**ndarray.size**
the total number of elements of the array. This is equal to the product of the elements of shape.

```python
a.size
```

###**ndarray.dtype**
an object describing the type of the elements in the array. One can create or specify dtypes using standard Python types. Additionally NumPy provides types of its own. numpy.int32, numpy.int16, and numpy.float64 are some examples. Remember that in int64, the 64 refers to the memory allocated to hold this character.

```python
a.dtype
```

###**ndarray.itemsize**
the size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.

```python
a.itemsize
```

###**type()**

the data type of the object.

```python
type(a)
```

the product operator * operates elementwise in NumPy arrays. The matrix product can be performed using the @ operator or the dot function or method:

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
A * B     # elementwise product
          # A = [[ a, b ],
          #      [ c, d ]]
          # B = [[ e, f ],
          #      [ g, h ]]
          # A * B = [[ ae, bf ],
          #          [ cg, dh ]]
          # a = 1, b = 2, c = 3, d = 4, e = 5, f = 6, g = 7, h = 8
          # ae = 1x5 = 5
          # bf = 2x6 = 12
          # cg = 3x7 = 21
          # dh = 4x8 = 32
```

```python
A @ B     # matrix product
          # A = [[ a, b ],
          #      [ c, d ]]
          # B = [[ e, f ],
          #      [ g, h ]]
          # A @ B = [[ ae + bg, af + bh ],
          #          [ ce + dg, cf + dh ]]
          # a = 1, b = 2, c = 3, d = 4, e = 5, f = 6, g = 7, h = 8
          # ae + bg = 1x5 + 2x7 = 19
          # af + bh = 1x6 + 2x8 = 22
          # ce + dg = 3x5 + 4x7 = 43
          # cf + dh = 3x6 + 4x8 = 50
```

```python
A.dot(B)  # another matrix product
          # A = [[ a, b ],
          #      [ c, d ]]
          # B = [[ e, f ],
          #      [ g, h ]]
          # A @ B = [[ ae + bg, af + bh ],
          #          [ ce + dg, cf + dh ]]
          # a = 1, b = 2, c = 3, d = 4, e = 5, f = 6, g = 7, h = 8
          # ae + bg = 1x5 + 2x7 = 19
          # af + bh = 1x6 + 2x8 = 22
          # ce + dg = 3x5 + 4x7 = 43
          # cf + dh = 3x6 + 4x8 = 50
```

+= and *=, act in place to modify an existing array rather than create a new one.

```python
a = np.ones((2, 3), dtype=int)
a
```

```python
rg = np.random.default_rng(1)  # create instance of default random number generator
b = rg.random((2, 3))
b
```

```python
# *= acts in place to modify an existing array, in our case array a=array([[1, 1, 1],
#      [1, 1, 1]]) will be multiplied by 3 elementwise rather than create a new one.
a *= 3
a
```

```python
# += acts in place to modify an existing array, in our case array a=array([[3, 3, 3],
#      [3, 3, 3]]) will be added elementwise to array b rather than create a new one.
b += a
b
```

By default, these operations apply to the array as though it were a list of numbers, regardless of its shape. However, by specifying the **axis** parameter (columns=0 and rows=1) you can apply an operation along the specified axis of an array:

```python
# this is to create our toy matrix
b = np.arange(24).reshape(4, 6)
b
```

```python
b.min(axis=1) # min of each row
```

```python
b.cumsum(axis=1)  # cumulative sum along each row
```

#Universal Functions
NumPy provides familiar mathematical functions such as sin, cos, and exp. In NumPy, these are called “universal functions” (**ufunc**). Within NumPy, these functions operate elementwise on an array, producing an array as output.

#Indexing, Slicing and Iterating
One-dimensional arrays can be indexed, sliced and iterated over, much like lists and other Python sequences.

```python
# this is to create our toy vector
a = np.arange(10)**3
a
```

```python
a[2] # get the third element of the array, remember that indexing starts with 0
```

```python
a[2:5] # get the third through fifth elements of the array (not including fifth)
```

```python
a[0:6:2] # from start to position 6, get every 2nd element (start:end:stepsize)
```

```python
a[::-1] # reversed array
```

```python
for i in a: # use for loop to access every element of the array
  print(i**(1 / 3.))
```

```python
for element in b.flat:
    print(element)
```
