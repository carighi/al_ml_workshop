#The Basics
NumPy’s main object is the homogeneous multidimensional array (ndarray). It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers. In NumPy dimensions are called *axes*.

##An example NumPy ndarray


```python
import numpy as np
a = np.arange(15).reshape(3, 5)
a
```

###**ndarray.ndim**
the number of axes (dimensions) of the array.



```python
a.ndim
```

###**ndarray.shape**
the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the number of axes, ndim.

```python
a.shape
```

###**ndarray.size**
the total number of elements of the array. This is equal to the product of the elements of shape.

```python
a.size
```

###**ndarray.dtype**
an object describing the type of the elements in the array. One can create or specify dtypes using standard Python types. Additionally NumPy provides types of its own. numpy.int32, numpy.int16, and numpy.float64 are some examples.

```python
a.dtype
```

###**ndarray.itemsize**
the size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.

```python
a.itemsize
```

###**type()**

the data type of the object.

```python
type(a)
```

**arange** function creates sequences of numbers.

**numpy.arange([start, ]stop, [step, ]dtype=None, *, like=None)**

Return evenly spaced values within a given interval.

Values are generated within the half-open interval [start, stop) (in other words, the interval including start but excluding stop).

```python
np.arange(10, 40, 5)
```

```python
np.arange(0, 3, 0.5) # it accepts float arguments
```

When arange is used with floating point arguments, it is generally not possible to predict the number of elements obtained, due to the finite floating point precision. For this reason, it is usually better to use the function linspace that receives as an argument the number of elements that we want, instead of the step:

```python
from numpy import pi
np.linspace(0, 2, 9)                   # 9 numbers from 0 to 2
```

```python
from numpy import pi
x = np.linspace(0, 2 * pi, 10)        # useful to evaluate function at lots of points
f = np.sin(x)
f
```

the product operator * operates elementwise in NumPy arrays. The matrix product can be performed using the @ operator or the dot function or method:

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
A * B     # elementwise product
          # A = [[ a, b ],
          #      [ c, d ]]
          # B = [[ e, f ],
          #      [ g, h ]]
          # A * B = [[ ae, bf ],
          #          [ cg, dh ]]
          # a = 1, b = 2, c = 3, d = 4, e = 5, f = 6, g = 7, h = 8
          # ae = 1x5 = 5
          # bf = 2x6 = 12
          # cg = 3x7 = 21
          # dh = 4x8 = 32
```

```python
A @ B     # matrix product
          # A = [[ a, b ],
          #      [ c, d ]]
          # B = [[ e, f ],
          #      [ g, h ]]
          # A @ B = [[ ae + bg, af + bh ],
          #          [ ce + dg, cf + dh ]]
          # a = 1, b = 2, c = 3, d = 4, e = 5, f = 6, g = 7, h = 8
          # ae + bg = 1x5 + 2x7 = 19
          # af + bh = 1x6 + 2x8 = 22
          # ce + dg = 3x5 + 4x7 = 43
          # cf + dh = 3x6 + 4x8 = 50
```

```python
A.dot(B)  # another matrix product
          # A = [[ a, b ],
          #      [ c, d ]]
          # B = [[ e, f ],
          #      [ g, h ]]
          # A @ B = [[ ae + bg, af + bh ],
          #          [ ce + dg, cf + dh ]]
          # a = 1, b = 2, c = 3, d = 4, e = 5, f = 6, g = 7, h = 8
          # ae + bg = 1x5 + 2x7 = 19
          # af + bh = 1x6 + 2x8 = 22
          # ce + dg = 3x5 + 4x7 = 43
          # cf + dh = 3x6 + 4x8 = 50
```

One-dimensional arrays can be indexed, sliced and iterated over, much like lists and other Python sequences.

```python
a[2] # get the third element of the array
```

```python
a[2:5] # get the third through fifth elements of the array
```

```python
a[0:6:2] # from start to position 6, get every 2nd element (start:end:stepsize)
```

```python
a[::-1] # reversed array
```

```python
for i in a: # use for loop to access every element of the array
  print(i**(1 / 3.))
```

When fewer indices are provided than the number of axes, the missing indices are considered complete slices:

```python
b[-1] # the last row. Equivalent to b[-1, :]
```

The dots (...) represent as many colons as needed to produce a complete indexing tuple. 

```python
b[1, ...] # same as b[1, :] or b[1]
```

```python
b[1, :]
```

```python
b[1]
```

```python
b[..., 2] # same as b[:, 2], third column
```

```python
b[:, 2]
```

Iterating over multidimensional arrays is done with respect to the first axis:



```python
for row in b:
  print(row)
```

However, if one wants to perform an operation on each element in the array, one can use the **flat** attribute which is an iterator over all the elements of the array:

```python
for element in b.flat:
    print(element)
```

#Shape Manipulation


## Changing the shape of an array
An array has a shape given by the number of elements along each axis:

```python
a = np.floor(10 * rg.random((3, 4)))
a
```

```python
a.shape
```

The shape of an array can be changed with various commands. Note that the following three commands all return a modified array, but do not change the original array:

```python
a.ravel()  # returns the array, flattened
```

```python
a.reshape(6, 2)  # returns the array with a modified shape
```

```python
a.T  # returns the array, transposed
```

```python
a.T.shape
```

```python
a.shape
```

The **reshape** function returns its argument with a modified shape, whereas the **ndarray.resize** method modifies the array itself:

```python
a
```

```python
a.resize((2,6))
a
```

If a dimension is given as -1 in a reshaping operation, the other dimensions are automatically calculated:

```python
a.reshape(3, -1)
```
